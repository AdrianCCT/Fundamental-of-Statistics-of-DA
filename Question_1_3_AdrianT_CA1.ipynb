```python
variance = np.var(df['Freshwater Withdrawals '])
print(f"Variance: {variance}")

#  I already know what standard deviation is from describe() code . I want to show that I am aware of different ways of calculating it (we alos can take square root of variance)
std_dev = np.std(df['Freshwater Withdrawals '])
mean = np.mean(df['Freshwater Withdrawals '])

cv = (std_dev / mean) * 100
print(f"Coefficient of Variation (CV): {cv:.2f}%")
```

```python
import statsmodels.api as sm
import scipy.stats as stats

# Histogram with KDE
plt.figure(figsize=(10, 6))
sns.histplot(df['Freshwater Withdrawals '], kde=True)
plt.title('Histogram of Freshwater Withdrawals with KDE')
plt.xlabel('Freshwater Withdrawals')
plt.ylabel('Frequency')

# QQ plot
plt.figure(figsize=(10, 6))
stats.probplot(df['Freshwater Withdrawals '], dist="norm", plot=plt)
plt.title('Q-Q Plot of Freshwater Withdrawals')
plt.show()

# Scattered plot wiht trendline
plt.figure(figsize=(8, 6))
plt.scatter(df['Year'], df['Freshwater Withdrawals '])
plt.xlabel('Year')
plt.ylabel('Freshwater Withdrawals ')
plt.title('Scatter Plot of Annual Freshwater Withdrawals with Trendline')
plt.grid(True)
x = df['Year']
y = df['Freshwater Withdrawals ']
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--")
plt.show()
```
